```python
import numpy as np
import os
import cv2
```

```python
directory_folder = '/content/dataset1_reyDistorsiones'
directory_tocopy_folder = '/content/dataset1_keypoints'
dir_images = os.listdir(directory_folder)
```

```python
def printImage(image):
  from matplotlib import pyplot as plt
  #Show the image with matplotlib
  plt.imshow(image)
  plt.show()
```

```python
fast = cv2.FastFeatureDetector_create() 
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()
```

```python
def detectKeypoints(image_name):  
  dir_image = f'{directory_folder}/{image_name}' 
  dir_tosave_image = f'{directory_tocopy_folder}/{image_name}'  
  image=cv2.imread(dir_image)
  # Detect keypoints with non max suppression
  keypoints = fast.detect(image, None)
  keypoints, brief_descriptor = brief.compute(image, keypoints)
  
  print (image_name)
  print ('Extracted '+str(len(keypoints))+' keypoints.')

  img=cv2.drawKeypoints(image,keypoints,image)
  cv2.imwrite(dir_tosave_image,image)
  printImage(image)
  print ('----------------')
  print ('')
```

```python
for image_name in dir_images:
  detectKeypoints(image_name)
```
