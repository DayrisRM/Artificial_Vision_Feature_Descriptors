```python
def printImage(image):
  from matplotlib import pyplot as plt
  #Show the image with matplotlib
  plt.imshow(image)
  plt.show()
```

```python
directory_folder = '/content/dataset2_reyCopia'
directory_tocopy_folder = '/content/dataset2_keypoints'
dir_images = os.listdir(directory_folder)
```

```python
# Create SURF object. You can specify params here or later.
surf = cv2.xfeatures2d.SURF_create()
```

```python
def detectKeypoints(image_name):  
  dir_image = f'{directory_folder}/{image_name}' 
  dir_tosave_image = f'{directory_tocopy_folder}/{image_name}'  
  image=cv2.imread(dir_image)  
  # Find keypoints and descriptors directly
  kpts, des = surf.detectAndCompute(image,None)  

  print (image_name)
  print ('Extracted '+str(len(kpts))+' keypoints.')

  img=cv2.drawKeypoints(image,kpts,image)
  cv2.imwrite(dir_tosave_image,image)
  printImage(image)
  print ('----------------')
  print ('')
```

```python
for image_name in dir_images:
  detectKeypoints(image_name)
```
